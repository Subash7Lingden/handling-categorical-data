### Count Or frequency encoding

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
train_set.head()                                                         
```

```python
columns =[1,3,5,6,7,8,9,13] # categorical columns
```

```python
train_set[columns]
```

```python
train_set=train_set[columns]
```

```python
### Creating our own columns
train_set.columns=['Employment','Degree', 'Status','Designation','family_job','Race','Sex', 'Country']
```

```python
train_set.head()
```

```python
for feature in train_set.columns[:]: # (every columns[:])
    print(feature,":",len(train_set[feature].unique()),'labels')
    
```

```python
## converting into dictionary
train_set['Country'].value_counts().to_dict()
```

- Here Indai is replaced by 100, cuba by 95, and so does other countries with there respective numbers

```python
## converting into dictionary
country_map =train_set['Country'].value_counts().to_dict()
```

```python
train_set['Country'] = train_set['Country'].map(country_map)
train_set.head(20)
```

#### Advantages
- It is easy to use
- We are not increasing feature space
 #### Disadvantages
- It will not be able to handle or provide the same weight if the frequencies are same.
    

### Target Guided Ordinal Encoding
- Ordering the labels according to the target
- Replace the labels by the joint probabilty of being 1 or 0

```python
import pandas as pd
df=pd.read_csv('titanic.csv', usecols=['Cabin','Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df.head(20)
```

```python
df['Cabin'].astype(str).str[0] # extracting the first latter
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df.head()
```

```python
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean()
```

- It gives the probability of survival of people in different cabins

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values() # sorting the values
```

```python
indexing the sorting values
df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels = df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
enumerate(ordinal_labels,0)
```

```python
ordinal_labels2 = {k:i for i, k in enumerate(ordinal_labels,0)} # this is dictionary comprehension
ordinal_labels2
```

- Based on the index value we assigned the ranking to the index
- Here based on the frequency of index D which have (D = 0.757576) values  it got the highets ranking.

- The dictionary is actually the dictionary comprehension. Here enumerate(ordinal_labels,0) will iterate thorugh each and every lebels, and the first value it assign is zero.
- When you do (ordinal_label ,0), the k basically indicates the ordinal lebals(index)['T', 'M', 'A', 'G', 'C', 'F', 'B', 'E', 'D'] and i indicates the labels that assigned values to labels(0,1,2,3,4,5,6,7,8)

```python
df['Cabin_ordinal_labels']=df['Cabin'].map(ordinal_labels2)
df.head()
```

### Probabilty ratio encoding

```python
df = pd.read_csv('titanic.csv', usecols=['Cabin','Survived'])
```

```python
df.head()
```

```python
### Replacing
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df.head(20)
```

```python
df['Cabin'].value_counts()
```

```python
df['Cabin'].unique() # it gives unique values of cabins
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df.head()
```

```python
df['Cabin'].unique() # This is the extraction of first letters from the  unique categories of cabin
```

```python
df.groupby(['Cabin'])['Survived'].mean() 
```

```python
prob_df = df.groupby(['Cabin'])['Survived'].mean() 
```

```python
prob_df = pd.DataFrame(prob_df)
prob_df
```

```python
## probability of people who died are
prob_df['died'] = 1 - prob_df['Survived']
```

```python
prob_df.head()
```

```python
# Calculating the probability ratio
prob_df['Probability_ratio'] = prob_df['Survived']/prob_df['died']
prob_df.head()
```

```python
# Converting the probability ratio into dictionary

prob_df['Probability_ratio'].to_dict()

```

```python
 probability_encoded= prob_df['Probability_ratio'].to_dict()
```

```python
df['Cabin_encoded'] = df['Cabin'].map(probability_encoded) ### assigning new category
df.head()
```

```python
df.head(20)
```

##### steps
- Probability of survived based in cabin categorical feature
- Probability of not survived  (1- probability of survived)
- Probability of survived/probability of dead
- Dictionary to map cabin with probability 
- Replace with the categorical feature
